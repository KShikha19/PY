1. Which 5 states have the highest total crash rates?

```python
top5_state = data.sort_values(by='total', ascending = False)
top5_state = top5_state.iloc[:5]
top5_state
```

Top 10 states with car crashes

```python
top10_state = data.sort_values(by='total', ascending = False)
top10_state = top10_state.iloc[:10]
top10_state
```

```python
sns.barplot(data=top10_state, x = 'total', y = 'abbrev', color='red')
plt.title('Top 10 states va total number of car crashes')
plt.ylabel('States')
plt.xlabel('Total crashes')
plt.show()
```

ND has maximum number of car crashes

```python
# plot to show state wise total accidents
sns.barplot(data=top5_state, x = 'abbrev', y = 'total')
plt.xlabel('States')
plt.ylabel('Total Accidents')
plt.show()
```

ND,SC and WW have max number of car crashes

2. Which 5 states have the lowest alcohol-related crash rates?

```python
low_alcohol_acc = data.sort_values(by = 'alcohol')
low_alchol_acc = low_alcohol_acc.iloc[:5]
low_alchol_acc
```

3. Which 5 states have the highest alcohol-related crash rates?


```python
top5alcohol_acc = data.sort_values(by = 'alcohol', ascending = False)
top5alcohol_acc = top5alcohol_acc.iloc[:5]
top5alcohol_acc
```

4. Top states having maxing number of accident due to alcohol


```python
df_alcohol = data[data['abbrev'].isin(top5_state['abbrev']) & data['abbrev'].isin(top5alcohol_acc['abbrev'])]
df_alcohol
```

```python
sns.barplot(data=df_alcohol, x='abbrev', y='total')
plt.xlabel('States')
plt.ylabel('Total Accidents')
plt.title('State Vs Alcohol Related Crashes')
plt.show()
```

ND and SC have same number of car crashes due to alcohol MT has less number of car crashes due to alchol than ND and SC

6. What is the corelation between total number of accident & speeding

```python
sns.scatterplot(data=data, x='total', y='speeding')
plt.show()
```

The number of speeding-related accidents tends to increase as the number of total accidents increases

```python
#correaltion coefficient between speeding and total number of accidents

corr_speeding_total  = data['total'].corr(data['speeding'])
print(corr_speeding_total*100)
```
