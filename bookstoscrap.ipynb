```python
import requests
from bs4 import BeautifulSoup

def book_finder():

  url  = 'https://books.toscrape.com/'
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'lxml')
  title_tag = soup.find('title')


  title = title_tag.text.strip()
  books = soup.find_all('li', class_ = 'col-xs-6 col-sm-4 col-md-3 col-lg-3')
  book_title_list = []
  book_price = []
  stock_availablity = []
  #book title list
  for book in books:
    if book:
      book_title = book.find('h3')
      book_title_list.append(book_title.text)
    else:
      book_title = 'N/A'
      book_title_list.append(book_title)

  #book price
    if book:
      price = book.find('p', class_ = 'price_color')
      book_price.append(price.text.strip())

    else:
      book_price.append('N/A')

    #stock availablity

    if book:
      stock = book.find('p',class_ = 'instock availability')
      stock_availablity.append(stock.text.strip())
    else:
      stock = 'N/A'
      stock_availablity.append(stock)
  print(f'Web Page Title is {title}')
  print(f'Extracted {len(book_title_list)} book postings')
  return {'Book Title' : book_title_list,
          'Book Price': book_price ,
          'Stock Availablity' : stock_availablity }
```

```python
book = book_finder()
book
```

```python
import pandas as pd
df = pd.DataFrame(book)
df.to_csv('book.csv',index=False)
display(df)
```
